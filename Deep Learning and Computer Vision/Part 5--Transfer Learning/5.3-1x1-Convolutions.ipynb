# 1x1 convolutions

1x1 Convolutions provide a flexible and efficient way to manipulate and transform features within a deep learning network. They allow for dimension reduction, non-linearity, feature combination, and increased model capacity, which can contribute to improved performance and representational power.

1x1 convolutions are commonly used in deep learning architectures for various purposes. Here are some of the main use cases:

    Dimension Reduction: 1x1 convolutions are often used to reduce the number of channels in a feature map, which helps reduce computational complexity and model size. By applying a 1x1 convolution with a smaller number of filters, the model can capture a compressed representation of the input.

    Non-linearity and Feature Combination: 1x1 convolutions introduce non-linearity to the network by applying element-wise activation functions. They can be used to combine features from different channels or feature maps, allowing the model to learn complex interactions between features.

    Increase Model Capacity: By adding 1x1 convolutions in the middle of the network, the model can increase its capacity without significantly increasing the computational cost. This can help capture more complex patterns and improve the overall representation power of the network.

    Network Inception/Inhibition: 1x1 convolutions can be used to control the flow of information within a network. By applying appropriate weights to the 1x1 filters, the model can emphasize or suppress certain features or channels, leading to better information flow and feature learning.

    Multi-Scale Feature Learning: 1x1 convolutions can be used in parallel branches to capture features at different scales or resolutions. By combining these branches, the model can effectively learn and utilize multi-scale information.

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D
from tensorflow.keras.utils import plot_model
```

```python
# Input shape
input_shape = (32, 32, 3)

# Input layer
inputs = Input(shape=input_shape)

# Apply a 1x1 convolution with 32 filters
conv_1x1 = Conv2D(32, (1, 1), activation='relu')(inputs)

# Apply a regular 3x3 convolution with 64 filters
conv_3x3 = Conv2D(64, (3, 3), activation='relu')(conv_1x1)

# Apply another 1x1 convolution with 128 filters
conv_1x1_2 = Conv2D(128, (1, 1), activation='relu')(conv_3x3)

# Create model
model = Model(inputs, conv_1x1_2)

# Display model summary
model.summary()
```

```python
plot_model(model,show_shapes = True, show_layer_names = True)
```
